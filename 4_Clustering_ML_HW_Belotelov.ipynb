```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
data = pd.read_excel('geo_comment.xlsx')
```

```python
data.head()
```

```python
data.shape
```

```python
data.comment_class.value_counts()
```

```python
data_district = data[['x', 'y']].sample(40000)
```

```python
scaler = StandardScaler()
```

```python
df_scaler = pd.DataFrame(scaler.fit_transform(data_district))
df_scaler.columns = data_district.columns
```

```python
df_scaler.head()
```

```python
kmeans = KMeans(n_clusters=4).fit(df_scaler[['x','y']])
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(df_scaler['x'], df_scaler['y'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
```

```python
X_scaled = scaler.fit_transform(data_district)
```

```python
model = DBSCAN()
prediction = model.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(12,10))
clusters = model.fit_predict(X_scaled)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='autumn', s=60, alpha=0.1)
plt.show()
```

```python
data_scaled = scaler.fit_transform(data[['x', 'y', 'comment_class', 'multiplier']])
```

```python
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(data_scaled)
    k_inertia.append(clf_kmeans.inertia_)
```

```python
plt.plot(ks, k_inertia)
plt.plot(ks, k_inertia ,'ro')
plt.show()
```

```python
diff = np.diff(k_inertia)
plt.plot(ks[1:], diff)
plt.show()
```

```python
diff_r = diff[1:] / diff[:-1]
plt.plot(ks[1:-1], diff_r)
plt.show()
```

```python
k_opt = ks[np.argmin(diff_r)+1]
k_opt
```
